```python
# Import PuLP
import pulp

# Create a Linear Programming Problem
prob = pulp.LpProblem("Maximize_Profit", pulp.LpMaximize)

# Decision Variables
x = pulp.LpVariable('Product_A', lowBound=0, cat='Continuous')
y = pulp.LpVariable('Product_B', lowBound=0, cat='Continuous')

# Objective Function
prob += 20*x + 30*y, "Total Profit"

# Constraints
prob += 2*x + 4*y <= 100, "Labor Constraint"
prob += 3*x + 2*y <= 90, "Material Constraint"

# Solve the problem
prob.solve()

# Results
print("Status:", pulp.LpStatus[prob.status])
print("Produce {} units of Product A".format(x.varValue))
print("Produce {} units of Product B".format(y.varValue))
print("Maximum Profit = $", pulp.value(prob.objective))
```

# Optimization Model: Maximizing Factory Profit

## Introduction
This project demonstrates how to solve a **business optimization problem** using **Linear Programming (LP)** in Python with the `PuLP` library.
The goal is to **maximize profit** for a factory producing multiple products under resource constraints like labor and material.


## Problem Data

| Product | Profit per unit ($) | Labor hours/unit | Material/unit |
|---------|------------------|-----------------|---------------|
| A       | 20               | 2               | 3             |
| B       | 30               | 4               | 2             |

**Available Resources:**  
- Labor hours = 100  
- Material = 90


## Linear Programming Formulation

**Decision Variables:**  
- x = units of Product A  
- y = units of Product B  

**Objective Function:**  
Maximize Profit = 20x + 30y  

**Constraints:**  
1. Labor: 2x + 4y ≤ 100  
2. Material: 3x + 2y ≤ 90  
3. Non-negativity: x ≥ 0, y ≥ 0


```python
# Import PuLP
import pulp

# Create a Linear Programming Problem
prob = pulp.LpProblem("Maximize_Profit", pulp.LpMaximize)

# Decision Variables
x = pulp.LpVariable('Product_A', lowBound=0, cat='Continuous')
y = pulp.LpVariable('Product_B', lowBound=0, cat='Continuous')

# Objective Function
prob += 20*x + 30*y, "Total Profit"

# Constraints
prob += 2*x + 4*y <= 100, "Labor Constraint"
prob += 3*x + 2*y <= 90, "Material Constraint"

# Solve the problem
prob.solve()

# Results
print("Status:", pulp.LpStatus[prob.status])
print("Produce {} units of Product A".format(x.varValue))
print("Produce {} units of Product B".format(y.varValue))
print("Maximum Profit = $", pulp.value(prob.objective))
```

## Results & Insights

**Status:** Optimal  
The solver found the best solution that maximizes profit while satisfying all constraints.

**Decision Variables:**  
- Produce 20 units of Product A  
- Produce 15 units of Product B  

**Maximum Profit:** $850

**Constraints Check:**  
- Labor used: 2*20 + 4*15 = 100 → fully used  
- Material used: 3*20 + 2*15 = 90 → fully used  

**Business Insights:**  
- Both products are produced to fully utilize available resources.  
- Product A has higher profit per labor hour, while Product B has higher profit per unit of material.  
- This balanced production plan maximizes total profit without exceeding available labor and material.


```python
import matplotlib.pyplot as plt

# Products and profit contribution
products = ['Product A', 'Product B']
profit = [20*x.varValue, 30*y.varValue]

# Plot
plt.bar(products, profit, color=['skyblue','orange'])
plt.ylabel('Profit ($)')
plt.title('Profit Contribution by Product')
plt.show()
```
